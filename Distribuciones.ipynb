```python
%matplotlib inline 

import numpy as np
import math
import matplotlib.pyplot as plt
import scipy as sp
import scipy.integrate
x=10
bi=[]
y=[]
bu=[]
acum=[]

#aleaorio gen
def numaleat():
    x0=7
    for i in range(x):
        mod=(223*x0+23)%829
        ui=float(mod)/829
        x0=mod
        y.append(ui)
    return y
plt.subplot(3,1,1)

def binomial():
    y=numaleat()
    for i in range(x):
        pr=(1-y[2])**(x-i)
        c=(math.factorial(x)/(math.factorial(i)*math.factorial(x-i)))*(y[2]**i)*pr
        bi.append(c)
    return bi
plt.plot(binomial(),color="orange",marker="*")
plt.show()

plt.subplot(3,1,2)
def acumubinomial():
    g=0
    for i in range(x):
        g=bi[i]+g
        acum.append(g)
    return acum
plt.plot(acumubinomial())
plt.show()


plt.subplot(3,1,3)
def invbin():
    y=numaleat()
    for i in range(x):
        for j in range(x):
            if y[i]<acum[j]:
                bu.append(j)
                break
    print("Valores Inversa Binomial",bu)
    return bu
plt.plot(invbin(),marker="x")
plt.grid()
plt.show()
```
